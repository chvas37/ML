```python
train2  = train.drop(['Ticket','Cabin'], axis=1)
train2
```

```python
#year = 2022 - train['Age']
#year
train2.loc[:, 'Age'] = train2["Age"].apply(lambda x:  2022-x)
train2['Age']
```

```python
train2 = train.rename({'PassengerId': 'НомерПассажира', 'Survived': 'Выжившие', 'Pclass': 'Пкласс', 'Name': 'Имя', 'Sex': 'Пол',	'Age':'Возраст',
                 'SibSp': 'Сибсп','Parch': 'жечь','Fare': 'транспрасходы',	'Embarked': 'места'}, axis=1)
train2
```

```python
train2['НомерПассажира'].fillna((train2['НомерПассажира'].mean()), inplace=True)
train2['Выжившие'].fillna((train2['Выжившие'].mean()), inplace=True)
train2['Пкласс'].fillna((train2['Пкласс'].mean()), inplace=True)
train2['Возраст'].fillna((train2['Возраст'].mean()), inplace=True)
train2['Сибсп'].fillna((train2['Сибсп'].mean()), inplace=True)
train2['жечь'].fillna((train2['жечь'].mean()), inplace=True)
train2['транспрасходы'].fillna((train2['транспрасходы'].mean()), inplace=True)
train2
```

```python
train2.to_csv('train2')
```
